```python
if False:
    # Start a new wandb run to track this experiment
    wandb_project = "LfI24"

    # Actual experiment
    run_name = "bl_enhanced_1_us"

    run = wandb.init(
        # set the wandb project where this run will be logged
        project=wandb_project,
        name=run_name)

    log = []
    modelname = "baseline"
    dataset_path = "data/dataset_us"
    results_path = "drive/Othercomputers/pc/train/results/colab"
    result_model_file = f"{results_path}/{run_name}.pth"
    result_model_final_file = f"{results_path}/{run_name}_final.pth"
    results_logs_file = f"{results_path}/{run_name}_log.json"

    # Number of data loader workers
    n_workers = n_cpu - 2
    if n_cpu <= 6:
        n_workers = n_cpu

    try:
        res = trainer.train_eval_model(
            model_name=modelname,
            data_dir=dataset_path,
            batch_size=128,
            epochs=30,
            momentum=0.9,
            lr=0.1,
            weight_decay=2e-05,
            useCosineAnnealingLR=True,
            warmup_epochs=5,
            warmup_lr_decay=0.01,
            label_smoothing=0.1,
            n_data_workers=n_workers,
            pin_memory=True,
            cuda_force=True,
            log=log,
            wandb_id=run.id,
            wandb_project=wandb_project,
            safe_model=True,
            safe_model_file=result_model_file,
            safe_model_file_final=result_model_final_file)

        misc.save_to_json_robust(res, results_logs_file)

    except KeyboardInterrupt:
        print("Interrupted by user.")
        misc.save_to_json_robust(log, f"{results_path}/{run_name}_log_unfinished.json")
```
